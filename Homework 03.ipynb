#### Задание 1. 

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.



Среднее арифметическое:

$$   \overline {x} = \frac { 100+80+75+77+89+33+45+25+65+17+30+24+57+55+70+75+65+84+90+150}{20} = \frac {$$

100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150

Так как вероятнсть попадания достаточно высокая и количество испытаний небольшое, то для расчет будем использовать формулу Бернулли для биноминальное распределения: 



$$ p(k=85)= C_n^k \cdot p^n \cdot q^{n-k} = C_{100}^{85}\cdot 0,8^{85} \cdot 0,2^{100-85} = \frac {100!}{85!(100-85)!}\cdot 0,8^{85} \cdot 0,2^{15}= \frac {86 \cdot 87 \cdot ... \cdot 100}{15!}\cdot 0,8^{85} \cdot 0,2^{15}$$



```python
import math

x = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# расчет среднего арифметического
s = 0
for i in range(len(x)):
    s += x[i]
x_average = s/len(x)
print('Среднее арифметическое: ', x_average)

# расчет среднего квадратичного отклонения
s = 0
for i in range(len(x)):
    s += (x[i]-x_average)**2
deviation = math.sqrt(s/len(x))
print('Среднее квадратичное отклонение: ', deviation)

# расчет дисперсии (смещенное отклонение)
offset_variance = deviation**2
print('Смещенная оценка дисперсии: ', offset_variance)


# расчет  несмещенного отклонения
s = 0
for i in range(len(x)):
    s += (x[i]-x_average)**2
unbiased_variance = s/(len(x)-1)
print('Несмещенная оценка дисперсии: ', unbiased_variance)
```

#### Задание 2. 


В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?



Для определения количества вариантов, которыми можно вытащит шары из коробки воспользуемся формулой сочетания:

$$C_n^k = \frac{n!}{k!(n - k)!}$$

При этом имеется 3 варианта извлечения 3 белых мячей из коробки:


Вариант 1 (2 белых из первой и 1 из второй).

Вероятность такого варианта расчитвается, как произведение вероятностей извлечь 2 белых шара из первой коробки и 1 белый шар из второй:

$$ p_1 = \frac{C_5^2*C_3^0}{C_8^2} \cdot \frac{C_5^1*C_7^3}{C_{12}^4}$$


```python
import numpy as np
import math

def C(n,k):
    return int(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))

p1 = (C(5,2)*C(3,0)/C(8,2)) * (C(5,1)*C(7,3)/C(12,4))
p1
```

Вариант 2 (1 белый из первой и 2 из второй).

Вероятность такого варианта рассчитвается, как произведение соответствующих вероятностей:

$$ p_2 = \frac{C_5^1*C_3^1}{C_8^2} \cdot \frac{C_5^2*C_7^2}{C_{12}^4}$$


```python
p2 = (C(5,1)*C(3,1)/C(8,2)) * (C(5,2)*C(7,2)/C(12,4))
p2
```

Вариант 3 (0 белых из первой и 3 из второй).

Вероятность такого варианта рассчитвается, как произведение соответствующих вероятностей:

$$ p_3 = \frac{C_5^0*C_3^2}{C_8^2} \cdot \frac{C_5^3*C_7^1}{C_{12}^4}$$


```python
p3 = (C(5,0)*C(3,2)/C(8,2)) * (C(5,3)*C(7,1)/C(12,4))
p3
```

Тогда вероятность извлечения 3 белых шаров в разных комбинациях будет равна сумме вероятностей:

$$ p = p_1+p_2+p_3 $$

```python
p=p1+p2+p3
p
```

#### Задание 3. 

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.


Для решения задачи воспользуемся формулой Байеса:

$$ P(b \mid a) = \frac{p(b) \cdot p(a \mid b)}{p(a)} $$

где b - это вероятность, что стреляет i-й стрелок, а а - это вероятность попадания спортсмена, которые у нас равновероятны и соответственно равны 1/3.

Тогда определим недостающую для нашей задачи вероятность попадания в мишень случайным спортсменом:

$$ p_a = p_1(b)\cdot p_1(a \mid b) + p_2(b)\cdot p_2(a \mid b)+p_3(b)\cdot p_3(a \mid b) $$

```python
p1 = 0.9
p2 = 0.8
p3 = 0.6

pa = 1/3*p1+1/3*p2+1/3*p3

print('Вероятность попадания спортсменами в мишень p(a): ', pa)
print('Вероятность того, что стрелял 1-й спортсмен p1(b|a): ', (1/3*p1)/pa)
print('Вероятность того, что стрелял 2-й спортсмен p2(b|a): ', (1/3*p2)/pa)
print('Вероятность того, что стрелял 3-й спортсмен p3(b|a): ', (1/3*p3)/pa)
```

#### Задание 4. 


В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?



Для решения задачи снова воспользуемся формулой Байеса:

$$ P(b \mid a) = \frac{p(b) \cdot p(a \mid b)}{p(a)} $$

где b - это вероятность, что студент с определенного факультета, а а - это вероятность того, что студент сдал сессию. Учитывая, что на факультете с учится столько же студентов, сколько на а и b вместе, то найдем недостающую нам вероятность сдачи сессии студентами:

$$ p_{ex} = p_a(b)\cdot p_a(a \mid b) + p_b(b)\cdot p_b(a \mid b)+p_c(b)\cdot p_c(a \mid b) $$

```python
pa = 0.8
pb = 0.7
pc = 0.9

p_ex = 1/4*pa+1/4*pb+1/2*pc

print('Вероятность успешной сдачи сессии студентами всех факультетов p(a): ', p_ex)
print('Вероятность того, что сдал сессию студент факультета а  pа(b|a): ', (1/4*pa)/p_ex)
print('Вероятность того, что сдал сессию студент факультета а  pа(b|a): ', (1/4*pb)/p_ex)
print('Вероятность того, что сдал сессию студент факультета а  pа(b|a): ', (1/2*pc)/p_ex)
```

#### Задание 5. 


Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?



Обозначим вероятности поломки детали соответствующими индексами:

$$p_1=0.1$$

$$p_2=0.2$$

$$p_3=0.25$$

Тогда:

а) Вероятность, что все детали одновременно выйдут строя равна:

$$ p = p_1 \cdot p_2 \cdot p_3 = 0.1 \cdot 0.2 \cdot 0.25 = 0.005$$

```python
# вероятность выхода детали из строя
p1 = 0.1
p2 = 0.2
p3 = 0.25

# вероятность сохранения работоспособности детали
q1 = 1 - p1
q2 = 1 - p2
q3 = 1 - p3

p = p1*p2*p3
p
```

b) Вероятность, что только две детали одновременно выйдут строя равна:

$$ p = p_1 \cdot p_2 \cdot q_3 + p_1 \cdot q_2 \cdot p_3 + q_1 \cdot p_2 \cdot p_3 = 0.08$$

```python
p = p1*p2*q3 + p1*q2*p3 + q1*p2*p3
p
```

c) Вероятность, что хотя бы одна деталь выйдет из строя, то есть будет событие противоположное тому, что все детали будут работать:

$$ p = 1 - q_1 \cdot q_2 \cdot q_3 = 0.46$$

```python
p = 1-q1*q2*q3
p
```

d) Вероятность, что одна или две детали выйдут из строя, это значит, что произойдет событие, равновероятное что не случится событие что все детали будут работать или все детали выйдут из строя:

$$ p = 1 - q_1 \cdot q_2 \cdot q_3 - p_1 \cdot p_2 \cdot p_3= 0.455$$

```python
p = 1-q1*q2*q3 - p1*p2*p3
p
```

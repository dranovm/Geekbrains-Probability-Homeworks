#### Задание 1. 

Из колоды в 52 карты извлекаются случайным образом 4 карты. 
a) Найти вероятность того, что все карты – крести. 
б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.


а) В колоде, состоящей из 52 карт 4 мастей будет 52/4=13 карт одной масти, в том числе крестей. 

Соответственно количество вариантов комбинаций 4 карт одной масти в колоде будет равно сочетанию 4 из 13:

$$ C_{13}^{4} = \frac{13!}{4!(13-4)!} = \frac{10 \cdot 11 \cdot 12 \cdot 13}{2 \cdot 3 \cdot 4} = 5 \cdot 11 \cdot 13 = 715$$

Теперь оценим сколько вообще вариантов комбинаций 4 карт в колоде из 52 карт:

$$ C_{52}^{4} = \frac{52!}{4!(52-4)!} = \frac{49 \cdot 50 \cdot 51 \cdot 52}{2 \cdot 3 \cdot 4} = 49 \cdot 25 \cdot 17 \cdot 13 = 270725$$

Таким образом, вероятность выбрать 4 крестовые карты:

$$ P = \frac{715}{270725}= 0,0026 = 0,26\% $$

Ответ: 0,26%

б) В колоде, состоящей из 52 карт 4 туза. Значит количество сочетаний какого-либо туза с тремя другими картами будет:


$$ 4 \cdot C_{51}^3 = 4 \cdot \frac {51!}{3!(51−3)!}= 4 \cdot \frac{49 \cdot 50 \cdot 51}{2 \cdot 3}= 4 \cdot 49 \cdot 25 \cdot 17 = 83300 $$

Обращаем внимание, что в нашей задаче мы в отличии от урока выбираем из 52-1, а не из 52-4 карт, так как тузов в нашей выборке может быть больше одного. Сочетание из 48 использовалось, если бы в задаче необходимо было требование найти 4 карты только с 1 тузом.
 
Как и выше общее количество комбинаций 4 карт в колоде из 52 карт:

$$ C_{52}^{4} = \frac{52!}{4!(52-4)!} = \frac{49 \cdot 50 \cdot 51 \cdot 52}{2 \cdot 3 \cdot 4} = 49 \cdot 25 \cdot 17 \cdot 13 = 270725$$
 
Таким образом, вероятность выбрать 4 крестовые карты:

$$ P = \frac{83300}{270725}= 0,3077 = 30,77\% $$

Ответ 30,77%


#### Задание 2. 


На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?




Вариантов нажатия одновременно 3 кнопок (без определенного порядка) из 10:



$$ C_{10}^{3} = \frac{10!}{3!(10-3)!} = \frac{8 \cdot 9 \cdot 10}{2 \cdot 3} = 4 \cdot 3 \cdot 10 = 120$$
 
 Ответ: 120

#### Задание 3. 

В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?
 

Вероятность первая извлеченная деталь будет окрашена составляет:

$$ P_1 = \frac{9}{15} = 0,6 $$

После того как будет вытащена одна окрашенная деталь, вероятность вытащить вторую окрашенную составит:

$$ P_2 = \frac{8}{14} = 0,57 $$

Вероятность вытащить третью окрашенную деталь при условии, что две предыдущие были окрашены  составит:

$$ P_3 = \frac{7}{13} = 0,54 $$

А вероятность вытащить одновременно три окрашенные детали составит:

$$ P = P_1 \cdot P_2 \cdot P_3 = 0,6 \cdot 0,57 \cdot 0,54 = 0,18 = 18 \% $$




#### Задание 4. 

В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?
 


Вероятность извлечения первого выигрышного билета будет 0,02:


$$ P_1 = \frac{2}{100} = 0,02 $$

После того как будет вытащен один лотерейный билет, вероятность вытащить второй составит:

$$ P_2 = \frac{1}{99} = 0,01 $$

Вероятность вытащить два выигрышных билета из 100, в которых 2 выигрышных составит:

$$ P = З_1 \cdot P_2 = 0,02 \cdot 0,01 = 0,0002 = 0,02 \% $$
